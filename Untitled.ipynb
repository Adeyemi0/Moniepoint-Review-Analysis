```python
from google_play_scraper import app, Sort, reviews_all
from app_store_scraper import AppStore
import pandas as pd
import numpy as np
import json, os, uuid
```

```python
g_reviews = reviews_all(
    "team.opay.pay",
    sleep_milliseconds=0,
    lang='en', # defaults to 'en'
    country='ng', # defaults to 'ng'
    sort=Sort.NEWEST, # defaults to Sort.NEWEST
)
```

```python
opay = pd.DataFrame(g_reviews)
opay.head()
```

```python
opay.columns
```

```python
a_reviews = AppStore('ng', 'opay-we-are-beyond-banking', '1463776084')
a_reviews.review()
```

```python
opay2 = pd.DataFrame(a_reviews.reviews)

# Check the first few rows of the DataFrame
opay2.head()
```

```python
opay2['developerResponse'][0]
```

```python
# Define a function to safely extract values from the dictionary
def extract_response_id(response):
    if isinstance(response, dict):  # Check if it's a dictionary
        return response.get('id', None)  # Use .get() to avoid KeyError
    return None

def extract_response_body(response):
    if isinstance(response, dict):  # Check if it's a dictionary
        return response.get('body', None)  # Use .get() to avoid KeyError
    return None

# Create new columns by applying the extraction functions
opay2['response_id'] = opay2['developerResponse'].apply(extract_response_id)
opay2['response_body'] = opay2['developerResponse'].apply(extract_response_body)

opay2.drop(columns=['developerResponse'], inplace=True)

opay2.head()
```

```python
opay2.columns
```

```python
opay = opay.rename(columns={
    'content': 'review',
    'score': 'rating',
    'at': 'date',
    'replyContent': 'response_body',  # Standardizing replyContent as response_body
    'reviewId': 'response_id'  # Standardizing reviewId as response_id
})

opay.head()
```

```python
opay = opay[['response_id', 'userName', 'review', 'rating', 'date', 'response_body']]
opay2 = opay2[['response_id', 'userName', 'review', 'rating', 'date', 'response_body']]

# Step 3: Add a column to indicate the source of the review
opay['source'] = 'Google Play Store'
opay2['source'] = 'Apple Store'

# Step 4: Append the two datasets
combined_reviews = pd.concat([opay, opay2], ignore_index=True)

# Display the merged DataFrame
combined_reviews.head()
```

```python
# Save the first 5 rows of combined_reviews to a CSV file
combined_reviews.to_csv('opay.csv', index=False)
```
